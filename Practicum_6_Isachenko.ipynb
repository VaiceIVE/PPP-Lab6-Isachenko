**Задание 1, 2**

```python
import pandas as pd
```

```python
df = pd.read_csv("chess.csv", delimiter=';|\t|,')
df = df.dropna()
df.columns = [col[1:-1] for col in df.columns]
df
```

```python
list_to_change = ["time_control", "end_time", "time_class", 	"rules", 	"gm_username", 	"white_username", "white_result", "black_username", "black_result"]
for name in list_to_change:
  df[name] = df[name].astype('string')
  df[name] = df[name].map(lambda x: x.strip('\"'))
df.sort_index()
df
```
